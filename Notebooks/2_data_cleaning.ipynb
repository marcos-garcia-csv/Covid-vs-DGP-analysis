```python
import pandas as pd
from IPython.core.interactiveshell import InteractiveShell
InteractiveShell.ast_node_interactivity = 'all'
```

```python
covid = pd.read_csv(r"E:\Datasets\Covid\Covid daily data.csv",parse_dates=['date'])
gdp_per_capita = pd.read_csv(r"E:\Datasets\GDP\GDP Per Capita.csv")
```

```python
#Dropping unnecessary columns

gdp_per_capita.drop(columns=[
    'Country Code',
    'Indicator Name',
    'Indicator Code'
],inplace=True
)
```

```python
#Handling missing values

covid.fillna(0,inplace=True)
```

```python
#Standarzing columns names and country names

gdp_per_capita = gdp_per_capita.rename(
    {
        'Country Name':'Country'
    },axis=1
)

covid.columns = covid.columns.str.replace("_"," ").str.title()
```

```python
covid_difference = pd.merge(
    covid,
    gdp_per_capita,
    on='Country',
    how='left'
)
covid_difference = covid_difference[covid_difference['Year'].isna()]
covid_difference['Country'].to_frame()

gdp_difference = pd.merge(
    gdp_per_capita,
    covid,
    on='Country',
    how='left'
)
gdp_difference = gdp_difference[gdp_difference['Active Cases'].isna()]
gdp_difference['Country'].to_frame()
```

```python
gdp_per_capita['Country'] = gdp_per_capita['Country'].replace(
{
   'Yemen, Rep':'Yemen',
   'Viet Nam':'Vietnam',
   'United Kingdom':'UK',
   'United States':'USA',
   "Cote d'Ivoire":"Cote D Ivoire",
   'St. Vincent and the Grenadines':'Saint Vincent And The Grenadines',
   'Congo, Rep.':'Congo',
   'Congo, Dem. Rep.':'Democratic Republic Of The Congo',
   'Bahamas, The':'Bahamas',
   'Gambia, The':'Gambia',
   'Egypt, Arab Rep.':'Egypt',
   'Micronesia, Fed. Sts.':'Micronesia',
   'Hong Kong SAR, China':'China',
   'Korea, Rep.':'South Korea',
   'Macao SAR, China':'Macao',
   'American Samoa':'Samoa'
})

covid['Country'] = covid['Country'].replace(
{
    'China Macao Sar':'Macao',
    'China Hong Kong Sar':'China',
    'Viet Nam':'Vietnam'
})
```

```python
covid
```

```python
gdp_per_capita = gdp_per_capita.pivot_table(
    values=['GDP per Capita'],
    index='Country',
    columns='Year',
    aggfunc='mean'
)
gdp_per_capita

covid['Year'] = covid['Date'].dt.year
covid = covid.pivot_table(
    values=['Cumulative Total Cases','Daily New Cases','Active Cases','Cumulative Total Deaths','Daily New Deaths'],
    index='Country',
    columns='Year',
    aggfunc="sum"
)
covid
```

```python
merged_df = covid.merge(
    gdp_per_capita,
    on=['Country'],
    how='inner'
)
```

```python
merged_df.fillna(0,inplace=True)
```

```python
merged_df.to_csv(r"E:\Datasets\cleaned_Data.csv",index=False)
```
