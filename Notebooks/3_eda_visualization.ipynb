```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv(r"E:\Datasets\cleaned_data.csv").set_index('Country')
```

```python
df.describe()
df.info()
```

```python
correlation_matrix = df.corr()
```

```python
# Create a heatmap using Seaborn
plt.figure(figsize=(10, 8))  # Set the figure size
sns.heatmap(correlation_matrix, annot=True, cmap="coolwarm", fmt=".2f")
plt.title("Correlation Heatmap")
plt.show()
```

```python
total_cases = {
    "2020": df['Cumulative Total Cases 2020'].sum(),
    "2021": df['Cumulative Total Cases 2021'].sum(),
    "2022": df['Cumulative Total Cases 2022'].sum()
}

total_deaths = {
    "2020": df['Cumulative Total Deaths 2020'].sum(),
    "2021": df['Cumulative Total Deaths 2021'].sum(),
    "2022": df['Cumulative Total Deaths 2022'].sum()
}

gdp_per_year = {
    "2020": df['GDP per Capita 2020'].mean(),
    "2021": df['GDP per Capita 2021'].mean(),
    "2022": df['GDP per Capita 2022'].mean()
}
```

```python
plt.Figure(figsize=(15,10))
plt.plot(total_cases.keys(),total_cases.values(),marker='o',label='Total Cases',linestyle='-',linewidth='2',color='orange')
plt.plot(total_deaths.keys(),total_deaths.values(),marker='s',label='Total Deaths',linestyle='--',color='red',linewidth='2')
plt.title('Covid-19 Total Cases & Deaths (2020-2022)')
plt.xlabel('Year')
plt.ylabel('Count')
plt.legend()
plt.show()
```

```python
sns.scatterplot(data=df,x='Cumulative Total Deaths 2020',y='GDP per Capita 2020',label='2020')
sns.scatterplot(data=df,x='Cumulative Total Deaths 2021',y='GDP per Capita 2021',label='2021')
sns.scatterplot(data=df,x='Cumulative Total Deaths 2022',y='GDP per Capita 2022',label='2022')
plt.xlabel('Cumulative Total Deaths')
plt.ylabel('GDP per Capita')
plt.legend(title='Year') 
plt.title('Covid-19 Deaths VS GDP per Capita (2020-2022)')
plt.show()
```

```python
sns.scatterplot(data=df,x='GDP per Capita 2020',y='Cumulative Total Deaths 2020',label='2020')
sns.scatterplot(data=df,x='GDP per Capita 2021',y='Cumulative Total Deaths 2021',label='2021')
sns.scatterplot(data=df,x='GDP per Capita 2022',y='Cumulative Total Deaths 2022',label='2022')
plt.xlabel('GDP per Capita')
plt.ylabel('Cumulative Total Deaths')
plt.title('GDP VS Covid-19 Cases (2020-2022)')
plt.legend(title='Year') 
plt.show()
```
